```python
import random
from random import randint
from IPython.display import clear_output
```

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':[1, 11]}
```

```python
def clear_display():
    clear_output()

def print_message(msg):
    print(msg)
    
def print_error_message(msg):
    print(msg)

def get_user_int_input(msg, options = []):
    user_input = input(msg)
    try:
        user_input = int(user_input)
    except ValueError:
        print_error_message(f"Please, provide a number.")
        get_user_int_input(msg, options)

    if int(user_input) not in options:
        print_error_message(f"Please, provide the right number.")
        get_user_int_input(msg, options)
    
    return user_input

def get_user_str_input(msg, options = []):
    user_input = input(msg)
    if user_input not in options:
        print_error_message(f"Please, provide the right option.")
    return user_input
```

```python
class Card:
    def __init__(self, suit, rank, value):
        self.suit = suit
        self.rank = rank
        self.value = value
        self.face_up = False

    def __str__(self):
        value = self.value if not self.rank == "Ace" else f"1 or 11"
        return f"Suit: {self.suit} & Rank: {self.rank} & Value: {value}" if self.face_up else f"Card is faced down."

    def flip(self):
        if self.face_up:
            self.face_up = False
        else:
            self.face_up = True
```

```python
class Deck:
    def __init__(self):
        self.cards = []

        for suit in suits:
            for rank in ranks:
                self.cards.append(Card(suit, rank, values[rank]))

    def shuffle(self):
        random.shuffle(self.cards)
    
    def get_top_card(self):
        return self.cards.pop(0)
```

```python
class Bank:

    class Bank_Account:

        def __init__(self, player):
            self.player = player
            self.balance = 0

    def __init__(self):
        self.bank_accounts = []

    def create_account(self, player):
        self.bank_accounts.append(self.Bank_Account(player))

    def get_player_account(self, player):
        for bank_account in self.bank_accounts:
            if bank_account.player == player:
                return bank_account

    def inquiry_balance(self, player):
        bank_account = self.get_player_account(player)
        return bank_account.balance

    def deposit(self, player, amount):
        bank_account = self.get_player_account(player)
        bank_account.balance += amount

    def with_draw(self, player, amount):
        bank_account = self.get_player_account(player)
        if bank_account.balance >= amount:
            bank_account.balance -= amount
        else:
            print_message("Amount is Too High!!")
```

```python
class Player:
    def __init__(self, name = ""):
        self.name = name
        self.cards = []

    def get_hand_value(self):
        total_hand_value = 0
        aces = []
        for card in self.cards:
            if card.face_up:
                if card.rank == "Ace":
                    aces.append(card)
                else:
                    total_hand_value += card.value

        for ace in aces:
            if ace.face_up:
                if total_hand_value+11 <= 21:
                    total_hand_value += ace.value[1]
                else:
                    total_hand_value += ace.value[0]
        return total_hand_value

    def get_str_cards(self):
        str_cards = ""
        for card in self.cards:
            str_cards += str(card)+"\n"
        return str_cards

    def check_black_jack(self):
        return (self.cards[0].value == 10 and self.cards[1].rank == "Ace") or (self.cards[1].value == 10 and self.cards[0].rank == "Ace")
```

```python
class Dealer(Player):
    def __init__(self):
        Player.__init__(self)
```

```python
class Black_Jack_Table:
    def __init__(self, deck, bank, dealer, players):
        self.deck = deck
        self.bank = bank
        self.dealer = dealer
        self.players = players
        self.bets = {}

    def shuffle_deck(self):
        self.deck.shuffle()

    def give_player_card(self, player):
        player.cards.append(self.deck.get_top_card())

    def give_dealer_card(self):
        self.dealer.cards.append(self.deck.get_top_card())

    def face_up_players_cards(self):
        for player in self.players:
            for card in player.cards:
                if not card.face_up:
                    card.flip()

    def face_up_player_cards(player):
        for card in player.cards:
            if not card.face_up:
                card.flip()

    def face_up_dealer_cards(self):
        for card in self.dealer.cards:
            if not card.face_up:
                card.flip()
    
    def show_table(self):
        print_message(f"Dealer's Cards:\n{self.dealer.get_str_cards()}\nTotal Value: {self.dealer.get_hand_value()}\n")

        for player in self.players:
            print_message(f"{player.name}'s Cards:\n{player.get_str_cards()}\nTotal Value: {player.get_hand_value()}\n")

    def refresh_table(self):
        clear_display()
        self.show_table()

    def get_players_bets(self):
        for player in self.players:
            player_balance = self.bank.inquiry_balance(player)
            if player_balance < 2:
                play_again = get_user_str_input(f"You ({player.name}) can't bet anymore. Do you wanna play for fun? (Y, N)", ["Y","N"])
                player_bet = 0
            else: 
                player_bet = get_user_int_input(f"{player.name} place a bet (balance: {}) (limit: 2 up to 500)", range(2,501))
            self.bets[player] = player_bet

    def set_up_table(self):
        print_message("Setting up table...")

        print_message("Shuffling deck...")
        self.shuffle_deck()

        black_jack = []
        for player in self.players:
            self.give_player_card(player)
            self.give_player_card(player)
            self.face_up_player_cards(player)
            if player.check_black_jack():
                black_jack.append(player)

        self.show_table()

    def player_turn(self, player):
        user_input = get_user_int_input(f"{player.name}: Type (1 or 2) --> 1: Hit, 2: Stand", [1,2])
        if user_input == 1:
            self.give_player_card(player)
            self.face_up_player_cards(player)
            self.refresh_table()
            print_message(f"{player.name.upper()} Hits!")
            if player.get_hand_value() > 21:
                print_message(f"{player.name.upper()} Busts!")
            else:
                self.player_turn(player)
        elif user_input == 2:
            self.refresh_table()
            print_message(f"{player.name.upper()} Stands.")
    
    def dealer_turn(self):
        self.face_up_dealer_cards()
        self.refresh_table()
        dealer_hand_value = self.dealer.get_hand_value()
        if dealer_hand_value in range(17,22):
            print_message("Dealer Stands.")
        elif dealer_hand_value > 21:
            print_message("Dealer Busts!")
        else:
            self.give_dealer_card()
            self.dealer_turn()

    def play_round(self):
        
        black_jack = []
        for player in self.players:
            print_message(f"{player.name.upper()}'s Turn!")
            black_jack = self.player_turn(player)
            if black_jack:
                black_jack.append(player)
        self.dealer_turn()

        winners = []
        round_result = "\n----->End Round Result<-----\n"
        for player in self.players:
            winner = "Loser"
            hand_value = player.get_hand_value()
            if player in black_jack or (hand_value <= 21 and hand_value > self.dealer.get_hand_value()):
                winner = "Winner"
                winners.append(player)
            elif hand_value == self.dealer.get_hand_value():
                winner = "Draw"
            round_result += f"{player.name}'s score: {hand_value} ({winner})\n"

        round_result += f"Dealer's score: {self.dealer.get_hand_value()}\n"

        print_message(round_result)
```

```python
print_message("Creating players...")
player1 = Player("Gustavo")
#player2 = Player("Amanda")
players = [player1]

bank = Bank()

print_message("Setting bank accounts...")
for player in players:
    bank.create_account(player)

bj_table = Black_Jack_Table(Deck(), bank, Dealer(), players)

bj_table.set_up_table()
bj_table.play_round()
```
